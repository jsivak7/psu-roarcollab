# NiOconv2x2x2 NCORE

$$Speedup = \frac{t(1)}{t(N)}$$

where $t(1)$ is the computational time for running the software using one core, and $t(N)$ is the computational time running the same software with N cores. (https://hpc-wiki.info/hpc/Scaling)

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = [5, 3]
plt.rcParams['font.size'] = 10
plt.rcParams['axes.linewidth'] = 1.25
```

```python
bc_sr = pd.read_csv('../data/NiOconv2x2x2/ncore_singlenode/bc_sapphirerapids.csv')
sc_cl = pd.read_csv('../data/NiOconv2x2x2/ncore_singlenode/sc_cascadelake.csv')
sc_il = pd.read_csv('../data/NiOconv2x2x2/ncore_singlenode/sc_icelake.csv')
```

```python
bc_sr
```

```python
for ncore in [1, 8, 16, 24, 32, 48, 64]:
    plt.plot(bc_sr.num_cores, bc_sr[f"ncore1"][0]/bc_sr[f"ncore{ncore}"], label=ncore, marker='+', alpha=0.8)
plt.plot([0, 64], [0, 64], color='black', linestyle='--')
plt.xlabel("# cores")
plt.ylabel("speedup")
plt.title("Basic Cores (SapphireRapids)", weight='bold')
plt.legend(title='NCORE', loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(alpha=0.1)
plt.tight_layout()
plt.savefig('../results/NiOconv2x2x2_ncore_bc_sapphirerapids.png')
```

```python
for ncore in [1, 8, 16, 24, 32, 48]:
    plt.plot(sc_cl.num_cores, sc_cl[f"ncore1"][0]/sc_cl[f"ncore{ncore}"], label=ncore, marker='+', alpha=0.8)
plt.plot([0, 48], [0, 48], color='black', linestyle='--')
plt.xlabel("# cores")
plt.ylabel("speedup")
plt.title("Standard Cores (CascadeLake)", weight='bold')
plt.legend(title='NCORE', loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(alpha=0.1)
plt.tight_layout()
plt.savefig('../results/NiOconv2x2x2_ncore_sc_cascadelake.png')
```

```python
for ncore in [1, 8, 16, 24, 32, 48]:
    plt.plot(sc_il.num_cores, sc_il[f"ncore1"][0]/sc_il[f"ncore{ncore}"], label=ncore, marker='+', alpha=0.8)
plt.plot([0, 48], [0, 48], color='black', linestyle='--')
plt.xlabel("# cores")
plt.ylabel("speedup")
plt.title("Standard Cores (IceLake)", weight='bold')
plt.legend(title='NCORE', loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid(alpha=0.1)
plt.tight_layout()
plt.savefig('../results/NiOconv2x2x2_ncore_sc_icelake.png')
```
